```python
khodro=khodro[::-1]
shapna=shapna[::-1]
up=up[::-1]
```

```python
khodro['last_day']=khodro.Close.shift(1)
up['last_day']=up.Close.shift(1)
shapna['last_day']=shapna.Close.shift(1)
```

```python
khodro['pnl']=khodro.Close - khodro.last_day
up['pnl']=up.Close - up.last_day
shapna['pnl']=shapna.Close - shapna.last_day
```

```python
new_df_shapna=np.array(shapna.pnl[::-1])
```

```python
new_df_up=np.array(up.pnl[::-1])
```

```python
num_days_khodro=len(khodro.index)
pnl=khodro.pnl
weakly_losses_khodro=[]
monthly_losses_khodro=[]
```

```python

for i in range(0,num_days_khodro):
    if i>=5 and pnl.iloc[i-5] > pnl.iloc[i]:
        weakly_losses_khodro.append(pnl.iloc[i]-pnl.iloc[i-5])
    if i>=20 and pnl.iloc[i-20] > pnl.iloc[i]:
        monthly_losses_khodro.append(pnl.iloc[i]-pnl.iloc[i-20])
plt.hist(weakly_losses_khodro,50)
plt.gca().set(title='weakly loss iran khodro')
```

```python
plt.hist(monthly_losses_khodro,50)
plt.gca().set(title='monthly loss iran khodro')
```
